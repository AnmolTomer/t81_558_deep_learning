````python
import base64
import os
import numpy as np
import pandas as pd
import requests

# This function submits an assignment.  You can submit an assignment as much as you like, only the final
# submission counts.  The paramaters are as follows:
# data - Pandas dataframe output.
# key - Your student key that was emailed to you.
# no - The assignment class number, should be 1 through 1.
# source_file - The full path to your Python or IPYNB file.  This must have "_class1" as part of its name.  
# .             The number must match your assignment number.  For example "_class2" for class assignment #2.
def submit(data,key,no,source_file=None):
    if source_file is None and '__file__' not in globals(): raise Exception('Must specify a filename when a Jupyter notebook.')
    if source_file is None: source_file = __file__
    suffix = '_class{}'.format(no)
    if suffix not in source_file: raise Exception('{} must be part of the filename.'.format(suffix))
    with open(source_file, "rb") as image_file:
        encoded_python = base64.b64encode(image_file.read()).decode('ascii')
    ext = os.path.splitext(source_file)[-1].lower()
    if ext not in ['.ipynb','.py']: raise Exception("Source file is {} must be .py or .ipynb".format(ext))
    r = requests.post("https://api.heatonresearch.com/assignment-submit",
        headers={'x-api-key':key}, json={'csv':base64.b64encode(data.to_csv(index=False).encode('ascii')).decode("ascii"),
        'assignment': no, 'ext':ext, 'py':encoded_python})
    if r.status_code == 200:
        print("Success: {}".format(r.text))
    else: print("Failure: {}".format(r.text))
````

````python
#Remove nan from length column, first check if NaN values exist
if (pd.isnull(df2['length']).values.any() == True):
    print(f"Length has NaN values ? {pd.isnull(df2['length']).values.any()}")
    print("Removing NaN values from df2['length']")
    length_median = df2['length'].median()
    df2['length'] = df2['length'].fillna(length_median)
else:
    print("No NaN values found!")
````

````python
#Check if still there are NaN values in df2['length'] as a proof of concept:
if (pd.isnull(df2['length']).values.any() == True):
    print(f"Length has NaN values ? {pd.isnull(df2['length']).values.any()}")
else:
    print("No NaN values found!")
````

````python
#Remove nan from height column, first check if NaN values exist
if (pd.isnull(df2['height']).values.any() == True):
    print(f"Height has NaN values ? {pd.isnull(df2['height']).values.any()}")
    print("Removing NaN values from df2['height']")
    length_median = df2['height'].median()
    df2['height'] = df2['height'].fillna(length_median)
else:
    print("No NaN values found!")
````

````python
if (pd.isnull(df2['height']).values.any() == True):
    print(f"Height has NaN values ? {pd.isnull(df2['height']).values.any()}")
else:
    print("No NaN values found in df2['height']!")
````

````python
#Convert df2['height]' to zscore

df2['height'] = zscore(df2['height'])
````

````python
df2 = df2.reindex(['height','max','number','length','cat2','item'],axis=1)
df2.head()
````

````python
# Check if max and number have any NaN values
if (pd.isnull(df2['max']).values.any() == True):
    print(f"max has NaN values ? {pd.isnull(df2['max']).values.any()}")
else:
    print("No NaN values found in df2['max']!")
    
if (pd.isnull(df2['number']).values.any() == True):
    print(f"number has NaN values ? {pd.isnull(df2['number']).values.any()}")
else:
    print("No NaN values found in df2['number']!")
````

````python
# Store max / number as ratio in ratio frame
ratio = df.apply(lambda x: x['max']/x['number'],axis=1)
display(ratio.head()) # Find column ratio = max/number
````

````python
df2.head() # Modify df2 instead of df original csv file
````

````python
df2.insert(4,"ratio",ratio) # insert frame ratio by name of ratio as 5th column
df2.head()
# df2 is almost close to what we need so further modifications will be done on a copy of df2 to avoid mess up
````

````python
df3 = df2
df3.head()
````

````python
cat2_dummy = pd.get_dummies(df3.cat2,prefix="cat2") # Create a column for every possible value

#Check if cat2_dummy has NaN

if (pd.isnull(cat2_dummy).values.any() == True):
    print(f"number has NaN values ? {pd.isnull(cat2_dummy).values.any()}")
else:
    print("No NaN values found in cat2_dummy!")
````

````python
#Replacing cat2 col with dummy variables
unique_cat2 = list(df3['cat2'].unique())
print(f"Number of unique cat2: {len(unique_cat2)}")
print(f"Unique cat2 list: {unique_cat2}")
````

````python
# There are 40 unique cat2 values, now we go to encoding dummy vars of col cat2

print(cat2_dummy.head())
````

````python
# If no NaN found add cat2 dummy variables to df3
if (pd.isnull(cat2_dummy).values.any() == False):
    df3 = pd.concat([df3,cat2_dummy],axis=1)
    display(df3.head())
````

````python
# Drop cat2 column

df3.drop('cat2',axis=1,inplace=True)
df3.head()
````

````python
# Do the same for item column

item_dummy = pd.get_dummies(df3.item,prefix="item") # Create a column for every possible value

#Check if cat2_dummy has NaN

if (pd.isnull(item_dummy).values.any() == True):
    print(f"Item has NaN values ? {pd.isnull(cat2_dummy).values.any()}")
else:
    print("No NaN values found in item_dummy!")
````

````python
#Replacing item col with dummy variables
unique_item = list(df3['item'].unique())
print(f"Number of unique item: {len(unique_item)}")
print(f"Unique item list: {unique_item}")
````

````python
# There are 31 unique item values, now we go to encoding dummy vars of item column

print(item_dummy.head())
````

````python
# If no NaN found add item_dummy variables to df3
if (pd.isnull(item_dummy).values.any() == False):
    df3 = pd.concat([df3,item_dummy],axis=1)
    display(df3.head())
````

````python
# Drop item column

df3.drop('item',axis=1,inplace=True)

df3.head()
````
